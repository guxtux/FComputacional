# Un modelo equivocado?

## Una fuente de error.

Se considera un tipo de error cuando consideramos un modelo que no est en congruencia con la fsica. Veamos el siguiente ejemplo.

### Planteando el problema.

La bicicleta es una forma muy eficiente de transporte, este es un hecho bien conocido por cualquier persona que la utliza.

Nuestro objetivo en este ejercicio es comprender los factores que determinan la velocidad mxima de una bicicleta y estimar la velocidad de un caso real.

Comenzaremos haciendo *caso omiso de la friccin*; tendremos que aadirlo al final, por supuesto, pero debemos primero entender cmo lidiar con el caso ms simple y sin friccin.

### Usando la mecnica de Newton.

La ecuacin de movimiento corresponde a la segunda ley de Newton, que escribimos de la forma:
$$ \begin{equation}
\dfrac{dv}{dt} = \dfrac{F}{m}
\label{EqNewton2} \tag{1}
\end{equation}
$$
donde $v$ es la velocidad, $m$ es la masa de la combinacin de la bicicleta-conductor, $t$ es el tiempo, y $F$ es la fuerza en la bicicleta que viene del esfuerzo del conductor (en este caso vamos a suponer que la bicicleta se mueve sobre un terreno plano).

Tratar correctamente a $F$ se complica por la mecnica de la bicicleta, ya que la fuerza ejercida por el ciclista se transmite a las ruedas por medio del plato, engranajes, cadena, etc.

Esto hace que sea muy difcil obtener una expresin exacta para $F$.

### Abordando el problema con otro enfoque.

Sin embargo, hay otra manera de abordar este problema que evita la necesidad de conocer la fuerza. Este enfoque alternativo implica la formulacin del problema en trminos de la potencia generada por el ciclista.

Estudios fisiolgicos de ciclistas de carreras han demostrado que estos atletas son capaces de producir una potencia de salida de aproximadamente $400$ Watts durante largos perodos de tiempo ($\sim 1$ hora).



Usando las ideas de trabajo-energa podemos reescribir la ec. (1) como:
$$
\begin{equation}
\dfrac{dE}{dt} = P
\label{EqPotencia} \tag{2}
\end{equation}
$$
donde $E$ es la energa total, $P$ es la potencia de salida del ciclista.

Para un trayecto plano la energa es totalmente cintica, es decir, $E = \frac{1}{2} m v^{2}$, y $dE/dt = m \, v (dv/dt)$, usando esto en la ec. (\ref{EqPotencia}), resulta:
$$
\begin{equation}
\dfrac{dv}{dt} = \dfrac{P}{m \, v}
\label{EqPotenciavel} \tag{3}
\end{equation}$$

Si $P$ es una constante, la ecuacin (\ref{EqPotenciavel}), se puede resolver de manera analtica, rearreglando trminos:

$$ 
\begin{equation}
\int_{v_{0}}^{v} v^{\prime} d v^{\prime} = \int_{0}^{t} \dfrac{P}{m} d t^{\prime}
\label{EqIntegral} \tag{4}
\end{equation}
$$
donde $v_{0}$ es la velocidad de la bicicleta en $t = 0$.

Integrando ambos lados de la ecuacin y resolviendo para $v$, tenemos que:
$$
\begin{equation}\label{Eqvres}
v = \sqrt{v_{0}^{2} + 2 \, P \dfrac{t}{m}} \tag{5}
\end{equation} $$

**La solucin es consistente con la fsica?**

Si bien esta es la solucin correcta de la ecuacin de movimiento (\ref{EqPotenciavel}), nuestro trabajo no concluye aqu: ya que predice que la velocidad se incrementar sin lmite para tiempos muy largos.

Este resultado *no es congruente* con la fsica que conocemos.

Vamos a corregir este resultado: cuando se generaliza el modelo se debe de *incluir el efecto de la resistencia del aire*

El nuevo trmino que vamos a aadir a la ecuacin de movimiento nos obliga a desarrollar una solucin numrica, as que con eso en mente se considera un tratamiento numrico de la ec. (\ref{EqPotenciavel}).

Comenzamos con la forma de diferencias finitas para estimar la derivada de la velocidad:

$$ \begin{equation}
\dfrac{dv}{dt} \simeq \dfrac{v_{i + 1} - v_{i}}{\Delta t}
\label{Eqderivada} \tag{6}
\end{equation}
$$
donde asumimos que $\Delta t$ es paso discreto pequeo, y $v_{i}$ es la velocidad al tiempo $t_{i} \equiv i \Delta t$.


Por lo que de la ecuacin (\ref{EqPotenciavel}):
$$
\begin{equation}
v_{i + 1} = v_{i} + \dfrac{P}{m \, v_{i}} \, \Delta \, t
\label{Eqveli+1} \tag{7}
\end{equation}
$$

Dada la velocidad en un tiempo $i$ (es decir, $v_{i}$), podemos usar (\ref{Eqveli+1}) para calcular un valor *aproximad* de la velocidad en el siguiente paso $v_{i+1}$.


Si conocemos la velocidad inicial $v_{0}$, podemos obtener $v_{1}$, $v_{2}$, y as sucesivamente.

### Proponemos un cdigo en python.

Ya contamos con suficientes elementos para proponer un cdigo en python, incluiremos una grfica del resultado.

```python
import matplotlib.pyplot as plt
from math import sqrt

t = []
v = []

dt = 1

potencia = 400
masa = 70
tmax = 200
nmax = tmax/dt

t.append(0)
v.append(4)

for i in range(int(tmax/dt)):
    ti = t[i-1] + dt
    vi = sqrt(v[i]**2 + (2 * potencia * dt)/masa)
    
    
    t.append(ti)
    v.append(vi)

plt.plot(v, "r-")
plt.xlabel("tiempo [s]")
plt.ylabel("velocidad m/s")
plt.show()
```

## Tenemos un resultado correcto?

Consideremos que:
1. Hemos desarrollado un algoritmo que resuelve el problema inicial.
2. Los valores son consistentes.
3. La grfica nos indica el comportamiento de la solucin.

An as: **debemos de considerar el problema como bien resuelto?**


## Considerando la friccin del aire.

### Segunda aproximacin.

La fuerza debida a la friccin puede aproximarse de manera inicial como:
$$
\begin{equation}
F_{a} \simeq - B_{1} \, v - B_{2} v^{2}
\label{EqFfriccion}
\end{equation}
$$
Para velocidades muy bajas, el primer trmino es el que domina, y el coeficiente $B_{1}$ se puede calcular para objetos con formas sencillas.

Para una velocidad razonable $v^{2}$ el trmino domina sobre los dems, pero $B_{2}$ no puede calcularse exactamente en objetos sencillos como una pelota de beisbol, menos para una bicicleta.

Podemos aproximar el valor de $B_{2}$ como sigue: Si un objeto se mueve a travs de la atmsfera, debe empujar fuera del camino el aire delante de l. 

La masa de aire movida en el tiempo $d t$ es 
$$m_{\text{aire}} \sim \rho A \: v \: dt$$

donde $\rho$ es la densidad del aire y $A$ el rea frontal del objeto.

Este aire tiene una velocidad de orden $v$, y por lo tanto, su energa cintica es $$E_{\text{aire}} \sim m_{\text{aire}} \: v^{2} / 2$$

Este es tambin el trabajo realizado por la fuerza de arrastre (la fuerza sobre el objeto debido a la resistencia del aire) en el tiempo $d t$, por lo tanto: $$F_{a} \: d t = E_{\text{aire}}$$

Al juntar estos resultados, encontramos que:
$$
\begin{align*}
F_{a} \simeq - C \: \rho \: A \: v^{2}
\end{align*}
$$

### Nueva expresin.

Incluyendo este trmino en la expresin para la velocidad:
$$
\begin{equation}
v_{i+1} = v_{i} + \dfrac{P}{m \: v_{i}} \Delta \: t - \dfrac{C \: \rho \: A \: v_{i}^{2}}{m} \: \Delta \: t
\label{Eqvelifriccion}
\end{equation}
$$
