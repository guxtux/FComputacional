# Ecuaciones Diferenciales Parciales.

Las ecuaciones diferenciales parciales (EDP) surgen en general, en relacin con fenmenos que tienen lugar en sistemas continuos, en los que las cantidades varan en el espacio y el tiempo.

Estos procesos son diversos: transporte de calor y masa, propagacin de ondas mecnicas, electromagnticas o mecnicas cunticas, etc.

Aparte de un nmero relativamente pequeo de casos simples, en los que las soluciones pueden expresarse en forma cerrada, es necesario recurrir a mtodos numricos para resolver las ecuaciones subyacentes.

Tpicamente, usando ciertos esquemas de discretizacin, la ecuacin diferencial se convierte en una ecuacin matricial, teniendo como incgnitas los valores de la solucin en los nodos de una malla espacio-tiempo.

Aunque la ecuacin matricial resultante puede resolverse en principio, por cualquiera de los mtodos generales (eliminacin gaussiana, factorizacin LU, etc.), la dimensin del sistema para problemas de inters prctico (del orden de miles) suele hacer inoperable tal enfoque.

El carcter local de las EDP consideradas (que contienen slo derivadas de bajo orden), as como el *carcter local* de los esquemas de discretizacin aplicados a los operadores diferenciales (implicando slo puntos de malla vecinos), genera con mayor frecuencia, que el sistema discretizado de ecuaciones sea el de una matriz dispersa: con elementos no nulos en slo unas pocas diagonales. Para tales matrices dispersas, existen tcnicas especiales de inversin.

Muchas de las EDP de importancia prctica en la fsica son ecuaciones de segundo orden, que contienen derivadas parciales de la funcin desconocida hasta el segundo orden y que suelen tener como variables independientes o bien coordenadas espaciales o espaciales y temporales.

## Forma general de las EDP.
Por simplicidad en problemas lineales en dos variables, las EDP tiene la forma genrica:

$$
\begin{align*}
a \: (x,y) \: \dfrac{\partial^{2} u}{\partial x^{2}} &+ b \: (x,y) \: \dfrac{\partial^{2} u}{\partial x \: \partial y} + c \: (x,y) \: \dfrac{\partial^{2} u}{\partial y^{2}} + \\
&+ d \: (x,y) \: \dfrac{\partial u}{\partial x} + e \: (x,y) \: \dfrac{\partial u}{\partial y} + \\
&+ g \: (x,y) \: u = f(x,y)
\end{align*}
$$

Donde $(x, y) \in D$ (dominio en el plano $x-y$) y la condicin $a^{2} (x, y) + b^{2} (x, y) + c^{2} (x, y)> 0$ debe mantenerse en todas partes en $D$.

Las EDP pueden clasificarse de acuerdo con la informacin de sus curvas de propagacin en:
1. Elpticas, si $b^{2} - 4 \: a\: c < 0$ para todo $(x,y) \in D$
2. Parablicas, si $b^{2} - 4 \: a \: c = 0$ para todo $(x,y) \in D$
3. Hiperblicas, si $b^{2} - 4 \: a \: c > 0$ para todo $(x,y) \in D$

# Ecuaciones Diferenciales Parciales.


## EDP Elptica.

Considera la ecuacin de Poisson:

$$
\begin{align*}
\dfrac{\partial^{2} \: u}{\partial x^{2}} + \dfrac{\partial^{2} \: u}{\partial y^{2}} = f(x, y)
\end{align*}
$$

En el caso de que no se incluya el trmino fuente $f(x, y)$, recuperamos la ecuacin de Laplace.

Las ecuaciones diferenciales parciales de tipo elptico estn dominadas por las derivadas homogneas de segundo orden, que aparecen en trminos que tienen el mismo signo.

Estas ecuaciones modelan de manera natural fenmenos estacionarios, por ejemplo tenemos: la ecuacin estacionaria de calor, la ecuacin de Poisson para el potencial electrosttico y la ecuacin de Schrdinger independiente del tiempo.

## EDP Parablica.


Ejemplos clsicos de las EDP parablicas son las ecuaciones de difusin y de calor:

$$
\begin{align*}
\dfrac{\partial u}{\partial x} - \dfrac{\partial}{\partial x} \left( D \: \dfrac{\partial u}{\partial x} \right) = 0 \hspace{1.5cm} \dfrac{\partial u}{\partial t} - K \: \dfrac{\partial^{2} u}{\partial x^{2}}  = f (x,y)
\end{align*}
$$

donde la evolucin est descrita en las derivadas de primer orden, $D$ es el coeficiente de difusin y $K > 0$ es la coeficiente de difusin trmica.

## EDP Hiperblica.

El ejemplo clsico para la EDP hiperblica es la \emph{ecuacin de onda}:

$$
\begin{align*}
\dfrac{\partial^{2} u}{\partial t^{2}}  - v^{2} \: \dfrac{\partial^{2} u}{\partial x^{2}} = 0
\end{align*}
$$

que incluye derivadas temporales y espaciales de segundo orden con signos contrarios, $v$ es la velocidad de fase de la onda.


# Condiciones de frontera.

En general, para obtener una solucin determinada de una EDP, que corresponde a una situacin fsica bien definida, tambin es necesario especificar las condiciones en donde se desarrolla, resultando ya sea en un problema de valores en la frontera o un problema de valores iniciales (*condiciones de Cauchy*).

Los problemas de los valores en la frontera estn tpicamente asociados con EDP elpticas y modelan fenmenos de equilibrio, para los cuales la evolucin en el tiempo es irrelevante.

Dada la ausencia de dependencia temporal, se busca una solucin en un estado estacionario $u(x, y)$ que satisface la ecuacin diferencial en un cierto dominio $D$, as como las condiciones de contorno asociadas. 


##  Condiciones de tipo Dirichlet.

Se especifican los valores solucin en la frontera:

$$
\begin{align*}
u (x, y) = u_{S}(x, y) \hspace{0.75cm} \mbox{ para } (x, y) \in S
\end{align*}
$$

## Condiciones de tipo Neumann.

Se definen las derivadas normales en la frontera:

$$
\begin{align*}
\dfrac{\partial u}{\partial \mathbf{n}} (x, y) = v_{S}(x, y) \hspace{0.75cm} \mbox{ para } (x, y) \in S
\end{align*}
$$

## Condiciones mixtas.

Involucran combinaciones lineales tanto de valores solucin de la funcin y de la derivada en la frontera:

$$
\begin{align*}
\alpha(x, y) \: u(x, y) + \beta(x, y) \: \dfrac{\partial u}{\partial \mathbf{n}} = \gamma(x, y) \hspace{0.5cm} \mbox{ para } (x, y) \in S
\end{align*}
$$

## Problemas de valores iniciales.

Los problemas de valor inicial (condiciones de Cauchy) naturalmente se asocian con ecuaciones parablicas o hiperblicas y tpicamente modelan fenmenos de propagacin.

Especficamente, basndose en el comportamiento espacial conocido de la solucin (y posiblemente tambin de su derivada temporal) en algn momento inicial $t_{0}$, la ecuacin diferencial gobierna la propagacin de la solucin $u(x, t)$ en el espacio y el tiempo.

## Solucin a los problemas.

Aunque el objetivo principal de ambos tipos de problemas es el clculo de la solucin en una red espacial, la solucin de estado estacionario para problemas con CDF se determina mediante un proceso numrico que converge simultneamente en todo el dominio $D$.

Mientras que en el caso de valores iniciales, la solucin de todos los puntos en el dominio espacial, se propagan de manera recursiva en el tiempo, iniciando la solucin a partir del instante $t_{0}$.
