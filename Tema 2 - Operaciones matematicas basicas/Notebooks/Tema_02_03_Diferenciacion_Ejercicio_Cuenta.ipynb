# Ejercicio a cuenta.

En un circuito eléctrico con voltaje $v(t)$ y una inductancia $L$, la primera ley de Kirchhoff se expresa como:
$$
\begin{align*}
v (t) = L \, \dfrac{d i}{d t} +  R \, i
\end{align*}
$$

donde $R$ es la resistencia e $i$ es la corriente. En el laboratorio se mide la corriente para distintos valores de $t$, obteniendo la siguiente tabla.

| t | 1.0  | 1.01 | 1.02 | 1.03 | 1.04 |
|--------|------|------|------|------|------|
| i      | 3.10 | 3.12 | 3.14 | 3.18 | 3.24 |

donde $t$ se mide en segundos, $i$ en Amperes, la inductancia $L$ es una constante de $0.98$ henries y $R$ es de $0.142$ Ohms.

**Aproxima el voltaje v(t) para cada valor de $t$**


Vamos a ocupar el **moduloDiferencias** para obtener el valor del voltaje en el extremo izquierdo, los puntos centrales y el extremo derecho de la tabla. Recordemos que usaremos las funciones que nos devuelven la aproximación de orden $O(h^{2})$.

## A revisar y ajustar.

Recuerda que el código devuelve las aproximaciones, pero le falta un "acabado" para mostrar la tabla y la correspondiente rutina de graficación. Por lo que deberás de completar esa parte y enviar tu notebook.

```python
from codigos.moduloDiferencias import dif_adelante_h2, dif_central_puntos, dif_atras_h2

def voltaje(derivada_i, corriente):
    valor = 0.98 * derivada_i + 0.142 * corriente
    return valor

tiempo = [1., 1.01, 1.02, 1.03, 1.04]
corriente = [3.10, 3.12, 3.14, 3.18, 3.24]
h = 0.1

derecha = dif_adelante_h2(corriente[0], corriente[1], corriente[2], h)
print(voltaje(derecha, corriente[0]))

for i in range(1, len(tiempo)-1):
    central = dif_central_puntos(corriente[i-1], corriente[i+1], h)
    print(voltaje(central, corriente[i]))

izquierda = dif_atras_h2(corriente[-1], corriente[-2], corriente[-3], h)
print(voltaje(izquierda, corriente[-1]))
```
