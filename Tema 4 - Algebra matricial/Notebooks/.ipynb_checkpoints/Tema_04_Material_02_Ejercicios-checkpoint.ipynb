## Ejercicio.

Se puede demostrar que las ecuaciones de movimiento de los bloques son:

$$
\begin{align*}
T_{1} + m_{1} \, a &= m_{1} \, g \, (\sin \theta - \mu_{1} \, \cos \theta) \\
- T_{1} + T_{2} + m_{2} \, a &= m_{2} \, g \, (\sin \theta - \mu_{2} \, \cos \theta) \\
- T_{2} + T_{3} + m_{3} \, a &= m_{3} \, g \, (\sin \theta - \mu_{3} \, \cos \theta) \\
- T_{-3} + m_{4} \, a &= - m_{4} \, g
\end{align*}
$$

donde $T_{i}$ denota las fuerzas de tensión en las cuerdas y $a$ es la aceleración del sistema. 

Calcula $a$ y $T_{i}$ si $\theta = 45^{\circ}$, $g = 9.82$ m/s2 y

$$
\mathbf{m} = [10 \quad 4 \quad \quad 5 \quad 6]^{T} \, \text{kg} \\[1em]
\mathbf{\mu} = [0.25 \quad 0.3 \quad 0.2]^{T} 
$$
