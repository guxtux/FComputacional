# Ejercicio con RK4.

Un sistema masa-resorte está en reposo hasta que se le aplica una fuerza $P (t)$, donde:
$$
\begin{align*}
P (t) = \begin{cases}
10 \, t \, \text{N} & \text{para } t < 2 \text{s} \\
20 \, \text{N} & \text{para } t \geq 2 \text{s}
\end{cases}
\end{align*}
$$



La EDO del movimiento resultante es:
$$
\begin{align*}
\ddot{y} = \dfrac{P (t)}{m} - \dfrac{k}{m} \, y
\end{align*}
$$

Resuelve lo siguiente:
1. Calcula el desplazamiento máximo de la masa.

Usa $m = 0.25$ kg y $k= 75$ N/m

```python
from Codigos.metodosDirectos import integraorden4
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    f = np.zeros(2)
    
    if x < 2:
        P = 10 * x
    else:
        P = 20
    
    f[0] = y[1]
    f[1] = (P/m) - (k/m) * y[0]
    return f

m = 0.25
k = 75
x = 0.0
xAlto = 5.0
y = np.array([0., 0.])
h = 0.01

X, Y = integraorden4(f, x, y, xAlto, h)

plt.plot(X, Y[:,0])
plt.title('Solución al ejercicio del sistema masa-resorte')
plt.xlabel('Tiempo [s]')
plt.ylabel('Desplazamiento [m]')
plt.xlim([0, 5])
plt.show()
```
