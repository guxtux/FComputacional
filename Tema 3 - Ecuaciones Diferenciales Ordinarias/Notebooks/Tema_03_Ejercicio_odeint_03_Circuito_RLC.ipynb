# Ejercicio 3 - Circuito RLC.

La corriente eléctrica de un circuito $RLC$ en serie:




Satisface la ecuación:
$$
\begin{align*}
L \, \dfrac{d i}{d t} + R \: i+ \dfrac{1}{C} \int_{0}^{t} i (t^{\prime}) d t^{\prime} +\dfrac{1}{C} \: q(0) = E(t), \hspace{0.3cm} t > 0 
\end{align*}
$$

Cuando el circuito se cierra en el instante $t = 0$, se tiene que $i = i(t)$ es la corriente, $R$ es la resistencia, $L, C, E$ vienen dadas por: $L = 200$ H, $C = 0.001$ F, $E(t) = 1$ V para $t > 0$.

## Condiciones iniciales.

Las condiciones iniciales son $q (0) = 0$ (carga inicial del condensador), $i (0) = 0$. 

# Por resolver.

Calcular la corriente para $0 \leq t \leq 5$ segundos y el factor de amortiguamiento y la frecuencia de oscilación del circuito $RLC$ para los siguientes valores de R:

1. $R = 0 \hspace{0.1cm} \Omega$
2. $R = 50 \hspace{0.1cm} \Omega$
3. $R = 100 \hspace{0.1cm} \Omega$
4. $R = 300 \hspace{0.1cm} \Omega$

# Resolviendo el ejercicio.

Si definimos
$$
\begin{align*}
q(t) = \int_{0}^{t^{\prime}} i(t^{\prime}) \: dt^{\prime}
\end{align*}
$$

derivando la expresión anterior:

$$
\begin{align*}
    \dfrac{d}{dt} \, q(t) = i(t), \hspace{1.5cm} q(0)=0
\end{align*}
$$

## Solución.

Sustituimos en la ecuación inicial, para reescribir
$$
\begin{align*}
\dfrac{d}{dt}i(t) = -\dfrac{R}{L} \: i(t) - \dfrac{1}{L \: C} \: q(t) + \dfrac{1}{L \: C} \: q(0) + \dfrac{E(t)}{L}, \: i(0) =  0 
\end{align*}
$$

La ecuación EDO2 inicial se transformó en un sistema de dos EDO1.

```python
%matplotlib inline

from numpy import zeros, array, linspace
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def F(y, t, R, L) : 
    C = 0.001
    E = 1.0
            
    F = zeros(2)
    F[0] = y[1]
    F[1] = -(R/L) * y[1] - (1.0/(L * C)) * y[0] + E/L
    
    return F

t = linspace(0.0, 7.0, 100)
y0 = array([0., 0.])


L = 200.0
R = [0., 50, 100, 500]

sol = odeint(F, y0, t)

# Hay que ajustar el código para que en un solo paso
# se genere la grafica con cada una de las respuestas del circuito RLC
# como la que se muestra a continuación
```
