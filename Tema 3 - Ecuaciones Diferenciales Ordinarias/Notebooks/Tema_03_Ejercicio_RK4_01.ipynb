Un flotador cónico se desliza libremente sobre una varilla vertical.



Al tocar el flotador, éste pierde su posición de equilibrio, y presenta un movimiento oscilante que se describe por la ecuación diferencial:
$$
\begin{align*}
\ddot{y} = g (1 - a \, y^{3})
\end{align*}
$$
donde $a = 16 \, m^{3}$ (que están determinadas por la densidad y dimensiones del flotador) y $g = 9.81$ m/s

Si el flotador se eleva a la posición $y = 0.1$ m y se libera, determina:
1. El período.
2. La amplitud.
de las oscilaciones.

```python
from Codigos.metodosDirectos import integraorden4
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    f = np.zeros(2)
    f[0] = y[1]
    f[1] = g * (1 - a * y[0]**3)
    return f

g = 9.81
a = 16
x = 0.0
xAlto = 10.0
y = np.array([0., 0.1])
h = 0.01

X, Y = integraorden4(f, x, y, xAlto, h)

plt.title('Solución al ejercicio del flotador')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [m]')
plt.xlim([0, 10])
```
